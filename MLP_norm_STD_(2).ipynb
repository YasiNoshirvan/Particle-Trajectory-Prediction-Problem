## Development output

```python
output1=development['x']
output2=development['y']
development_output = pd.concat([output1, output2], axis=1)
development_output
```

## Development input

```python
development_input =development.drop(columns=['x','y'],axis=1)
development_input
```

# **Feature selection process**

## Droping Columns

```python
rf = RandomForestRegressor(n_estimators=100,random_state=42).fit(development_input, development_output)
```

```python
featureImportance=rf.feature_importances_
```

```python
featureImportance
```

```python
x = []
for i in range(90):
  x.append(i)
```

```python
feature_names=np.array(development.columns)[2:]
feature_names
```

```python
sorted_feature_importance=sorted(zip(featureImportance, x), reverse=True)
sorted_feature_importance
```

```python
selected_features = sorted([row[1] for row in sorted_feature_importance[: 60]])

print(selected_features)
```

```python
evaluation_withOutID=evaluation.drop(columns=['Id'],axis=1)
```

```python
columns_to_keep = selected_features

remained_development=development_input.iloc[:, columns_to_keep]

remained_evaluation=evaluation_withOutID.iloc[:, columns_to_keep]
```

# **Normazlization and Standardization**

## Normalizing development input

```python
normalized_dev_input = normalize(remained_development, norm='l2')
normalized_dev_input
```

## Standardizing the normalized development input




```python
standardized_data = (normalized_dev_input - np.mean(normalized_dev_input, axis=0)) / np.std(normalized_dev_input, axis=0)
standard_input = pd.DataFrame(standardized_data, columns=remained_development.columns)
standard_input
```

## Showing normalized and standardized development data

```python
norm_development=pd.concat([development_output,standard_input],axis=1)
norm_development
```

# **Build Neural network model**

##  Splitting Trainset and Testset

```python
input_train, input_test, output_train, output_test = train_test_split(standard_input,development_output, test_size=0.01)
```

## Create the MLP Regressor On Train Data and Calculate accuracy metrics

```python
mlp_regressor = MLPRegressor(hidden_layer_sizes=(300,255), max_iter=400,verbose=True)
```

## Train the model

```python
mlp_regressor.fit(input_train, output_train)
```

## Predict on the test set

```python
predictions = mlp_regressor.predict(input_test)
```

## Evaluation the accuracy of fitted model

```python
mse = mean_squared_error(output_test, predictions)
print(f'Mean Squared Error: {mse}')

rmse = mean_squared_error(output_test, predictions, squared=False)
print(f'Root Mean Squared Error: {rmse}')

r2 = r2_score(output_test, predictions)
print(f'R-squared: {r2}')
```
